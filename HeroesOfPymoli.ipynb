```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
count = purchase_data.SN.nunique()
```

```python
# Display the total number of players
TotalCount = pd.DataFrame({"Total Number":[count]})
TotalCount
```

```python
uniqueitem = purchase_data["Item Name"].nunique()
avg = purchase_data.Price.mean()
purchase = (purchase_data["Purchase ID"]).count()
revenue = purchase_data.Price.sum()

Total = pd.DataFrame(
        { "Number of Unique Items":[uniqueitem],
          "Average Price": [avg],
          "Number of Purchases": [purchase],
          "Total Revenue": [revenue]  
        })


Total['Average Price'] = Total['Average Price'].map("${0:,.2f}".format)
Total['Total Revenue'] = Total['Total Revenue'].map("${0:,.2f}".format)
Total
```

```python
gender_total = purchase_data.groupby('Gender')['SN'].nunique()
gender_percentage = gender_total/gender_total.sum() * 100
gender_percentage
Final = pd.DataFrame({'Total Count': gender_total, 'Percentage of Players': gender_percentage}) 
                    
Final = Final.sort_values("Total Count",ascending =[False])
Final['Percentage of Players'] = Final['Percentage of Players'].map("{0:,.2f}%".format)
Final
```

```python
total_purchase = purchase_data.groupby('Gender')['Price'].count()

total_average = purchase_data.groupby('Gender')['Price'].mean()

total_value= purchase_data.groupby('Gender')['Price'].sum()

average_person = total_value/gender_total
Final1 = pd.DataFrame({'Purchase Count': total_purchase,
                      'Average Purchase price': total_average,
                       'Total Purchase Value': total_value,
                       'Avg Total Purchase per Person': average_person
                      })
Final1['Average Purchase price'] = Final1['Average Purchase price'].map("${0:,.2f}".format)
Final1['Total Purchase Value'] = Final1['Total Purchase Value'].map("${0:,.2f}".format)
Final1['Avg Total Purchase per Person'] = Final1['Avg Total Purchase per Person'].map("${0:,.2f}".format)
Final1
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
labels = ["<10","10 to 14","15 to 19","20 to 24", "25 to 29", "30 to 34", "35 to 39", "40+"]
pd.cut(purchase_data["Age"], bins, labels=labels).head()
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=labels)
#purchase_data.head()
```

```python
age_group = purchase_data.groupby("Age Group")

age_unique = age_group.SN.nunique()

age_percent = age_unique/age_unique.sum() * 100
#age_percent
Final2 = pd.DataFrame({'Total Count': age_unique,
                       'Percentage of Players': age_percent})
Final2['Percentage of Players'] = Final2['Percentage of Players'].map("{0:,.2f}%".format)
Final2
```

```python
age_count = age_group.SN.count()
tot_vol = age_group.Price.sum()
avg_tot = tot_vol/age_count
avg_purchase = tot_vol/age_unique

Final3 = pd.DataFrame({'Purchase Count': age_count,
                    'Average Purchase Price': avg_tot,
                    'Total Purchase Value': tot_vol,
                    'Avg Total Purchase per person': avg_purchase
                   })
Final3['Average Purchase Price'] = Final3['Average Purchase Price'].map("${0:,.2f}".format)
Final3['Total Purchase Value'] = Final3['Total Purchase Value'].map("${0:,.2f}".format)
Final3['Avg Total Purchase per person'] = Final3['Avg Total Purchase per person'].map("${0:,.2f}".format)
Final3
#print(purchase_count)
```

```python
sn_purchase = purchase_data.groupby('SN')['Price'].count()
#sn_purchase
sn_tot = purchase_data.groupby('SN')['Price'].sum()
sn_avg = sn_tot/sn_purchase 
Final4 = pd.DataFrame({'Purchase Count': sn_purchase,
                    'Average Purchase Price': sn_avg,
                    'Total Purchase Value': sn_tot
                   })
Final4 = Final4.sort_values('Total Purchase Value',ascending =[False])
Final4['Average Purchase Price'] = Final4['Average Purchase Price'].map("${0:,.2f}".format)
Final4['Total Purchase Value'] = Final4['Total Purchase Value'].map("${0:,.2f}".format)
Final4.head()
```

```python
Item_count = purchase_data.groupby(['Item ID','Item Name'])['Item Name'].count()
Item_sum   = purchase_data.groupby(['Item ID','Item Name'])['Price'].sum()
Price      = Item_sum/Item_count

Final5 = pd.DataFrame({'Purchase Count': Item_count,
                       'Price': Price,
                       'Total Purchase Value': Item_sum
                       })
Final5 = Final5.sort_values('Purchase Count',ascending =[False])
Final5['Price'] = Final5['Price'].map("${0:,.2f}".format)
Final5['Total Purchase Value'] = Final5['Total Purchase Value'].map("${0:,.2f}".format)
Final5.head()
```
